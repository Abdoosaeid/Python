# Introducing the pyplot interface 

```python
import matplotlib.pyplot as plt 
```

```python
fig , ax = plt.subplots() # 
plt.show()
```

- The plt-dot-subplots command, when called without any inputs, creates two different objects: a Figure object and an Axes object.
    - The Figure object **->** container that holds everything that you see on the page.
    - The Axes **->** the part of the page that holds the data. 

# Adding data to axes

```python
seattle_weather = {
    "MONTH":['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'],
    "MLY-TAVG-NORMAL":[42.1,43.4,46.6,50.5,56.0,61.0,65.9,66.5,61.6,53.3,46.2,41.1]
}
```

```python
fig, ax = plt.subplots()
ax.plot(seattle_weather["MONTH"],seattle_weather["MLY-TAVG-NORMAL"])
plt.show()
```

# Adding more data

```python
austin_weather = {
    "MONTH": ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    "MLY-TAVG-NORMAL": [52.3, 55.1, 60.2, 66.5, 74.8, 80.5, 84.2, 85.6, 78.4, 69.9, 59.8, 52.5]
}
```

```python
fig, ax = plt. subplots()
ax.plot(seattle_weather ["MONTH"], seattle_weather ["MLY-TAVG-NORMAL"])
ax.plot(austin_weather ["MONTH"], austin_weather ["MLY-TAVG-NORMAL"])
plt.show()
```

# Customizing your plots

## Adding markers 

```python
fig, ax = plt. subplots()
ax.plot(seattle_weather ["MONTH"], seattle_weather ["MLY-TAVG-NORMAL"],marker='o')
plt.show()
```

```python
fig, ax = plt. subplots()
ax.plot(seattle_weather ["MONTH"], seattle_weather ["MLY-TAVG-NORMAL"],marker='v')
plt.show()
```

### [matplotlib.markers](https://matplotlib.org/stable/api/markers_api.html)

## Setting the linestyle

```python
fig, ax = plt. subplots()
ax.plot(seattle_weather ["MONTH"], seattle_weather ["MLY-TAVG-NORMAL"],marker='v',linestyle="--")
plt.show()
```

```python
fig, ax = plt. subplots()
ax.plot(seattle_weather ["MONTH"], seattle_weather ["MLY-TAVG-NORMAL"],marker='v',linestyle="None")
plt.show()
```

# Choosing color 

```python
fig, ax = plt. subplots()
ax.plot(seattle_weather ["MONTH"], seattle_weather ["MLY-TAVG-NORMAL"],marker='v',linestyle="--",color="r")
plt.show()
```

# Customizing the axes labels

```python
fig, ax = plt. subplots()
ax.plot(seattle_weather ["MONTH"], seattle_weather ["MLY-TAVG-NORMAL"],marker='v',linestyle="--",color="r")
ax.set_xlabel("Time (months)")
plt.show()
```

```python
fig, ax = plt. subplots()
ax.plot(seattle_weather ["MONTH"], seattle_weather ["MLY-TAVG-NORMAL"],marker='v',linestyle="--",color="r")
ax.set_xlabel("Time (months)") 
ax.set_ylabel("Average temperature (Fahrenheit degrees)")
ax.set_title("Weather in Seattle")
plt.show()
```
